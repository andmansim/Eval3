#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
# your code here
fig, ax = plt.subplots(1,3, sharey=True)

columnas = ['Minutes Lightly Active', 'Minutes Very Active', 'Minutes Sedentary' ]
for i, j in enumerate(columnas):
    ax[i].scatter(x=fitbit[j], y=fitbit.Steps)
print(['Minutes Lightly Active', 'Minutes Very Active', 'Minutes Sedentary' ])
    
```
